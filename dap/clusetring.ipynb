```python
import pandas as pd
import numpy as np
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import LabelEncoder


```

```python
label_encoder = LabelEncoder()
df['Species'] = label_encoder.fit_transform(df['Species'])  # Correct column name to 'Species' # Drop 'Id' column before scaling
```

```python
features = ['SepalLengthCm', 'SepalWidthCm', 'PetalLengthCm', 'PetalWidthCm']
X = df[features].values

# Standardizing the features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

```

```python
while True:
	try:
		n_clusters = int(input("Enter the number of clusters for KMeans: "))
		break
	except ValueError:
		print("Invalid input. Please enter a valid integer.")

kmeans = KMeans(n_clusters, random_state=42)
clusters = kmeans.fit_predict(X_scaled)
```

```python
# Add the cluster assignments to the DataFrame
df['Cluster'] = clusters
plt.figure(figsize=(18, 8))
plt.subplot(1, 2, 1)
scatter = plt.scatter(df['PetalLengthCm'], df['PetalWidthCm'], c=df['Cluster'], cmap='viridis')
plt.title('K-means Clustering ')
plt.xlabel('Petal Length (cm)')
plt.ylabel('Petal Width (cm)')
plt.legend(handles=scatter.legend_elements()[0], labels=['Cluster 0', 'Cluster 1', 'Cluster 2'])
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Get number of rolls from user
num_rolls = int(input("Enter the number of die rolls: "))

# Simulate rolling a fair six-sided die
rolls = np.random.randint(1, 7, size=num_rolls)

# Create histogram of die roll outcomes
plt.figure(figsize=(8, 6))
plt.hist(rolls, bins=range(1, 8), align='left', rwidth=0.8, color='skyblue', edgecolor='black')
plt.title(f'Histogram of {num_rolls} Die Rolls')
plt.xlabel('Die Outcome')
plt.ylabel('Frequency')
plt.xticks(range(1, 7))
plt.grid(axis='y', alpha=0.75)
plt.show()

print(f"Simulated {num_rolls} rolls.")
```
